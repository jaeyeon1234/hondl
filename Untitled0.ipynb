```python
import numpy as np
import matplotlib.pyplot as plt

# 1. numpy.random.randn 함수를 이용해 4행 3열의 배열을 생성하세요.
np.random.seed(42)
array_a = np.random.randn(4, 3)
print("1. 4행 3열 배열 (array_a): ")
print(array_a)

# 2. 생성된 배열에서 두 번째 행부터 네 번째 행까지의 세 번째 열 값을 출력하세요.
slice_result = array_a[1:4, 2]
print("2. 두 번째 행부터 네 번째 행까지의 세 번째 열 값: ")
print(slice_result)

# 3. 전체 배열에서 양수 값만 추출하여 새로운 배열로 만드세요.
# 조건 필터링: array_a > 0
positive_array = array_a[array_a > 0]
print("3. 양수 값만 추출한 배열 (positive_array)")
print(positive_array)

# 4. 위에서 만든 양수 배열을 깊은 복사하여 새로운 배열을 만드세요.
sorted_array = positive_array.copy()
print("4. 깊은 복사된 배열 (sorted_array)")
print(sorted_array)

# 5. 복사한 배열을 오름차순으로 정렬하세요.
sorted_array = np.sort(sorted_array)
print("5. 오름차순으로 정렬된 배열 (sorted_array)")
print(sorted_array)

# 6. 원본 양수 배열과 정렬된 배열을 막대그래프로 시각화하세요.
# 원본 양수 배열 시각화
plt.bar(range(len(positive_array)), positive_array, color='skyblue')
plt.title('Original Positive Array')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()

# 정렬된 배열 시각화
plt.bar(range(len(sorted_array)), sorted_array, color='lightcoral')
plt.title('Sorted Positive Array')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```
